# **TRAYENDO LA GAMA DE COLORES ROJO**- **HISTOGRAMA Y DIAGRAMA CIRCULAR**

```python
imagenrojo=cv.cvtColor(imagen,cv.COLOR_BGR2RGB)
imagenrojo=np.ravel(imagenrojo)
imagenrojo,len(imagenrojo),np.shape(imagenrojo),np.shape(imagenrojo)
ROJOSclaros=imagenrojo[(imagenrojo>=170) & (imagenrojo<=255)]
ROJOSoscuros=imagenrojo[(imagenrojo>=0) & (imagenrojo<85)]
ROJOSmedios=imagenrojo[(imagenrojo>=85) & (imagenrojo<170)]
len(ROJOSclaros),len(ROJOSoscuros),len(ROJOSmedios)

plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
plt.bar(np.array(["Claro","Oscuro","Medio"]),np.array([len(ROJOSclaros),len(ROJOSoscuros),len(ROJOSmedios)])/len(imagenrojo))
plt.subplot(1,2,2)
plt.pie(np.array([len(ROJOSclaros),len(ROJOSoscuros),len(ROJOSmedios)])/len(imagenrojo), labels = np.array(["Claro","Oscuro","Medio"]))
plt.show()
```

```python
vArticulos=np.unique(df_excel["NombreArticulo"])
vIDart=np.unique(df_excel["ID"])
j=0
tTCompras=0
lTCompras=[]
lCon=[]
precioArt=[]
while(j < len(vArticulos)):
    #print("=========================================")
    mCompras=0
    tCompras=0
    con=0
    for i in range(len(df_excel["NombreArticulo"])):
        if(df_excel["NombreArticulo"][i]==vArticulos[j]):
            tCompras+=df_excel["Precio"][i]
            con+=1
        precioArt.append(df_excel["Precio"][i])
    tTCompras+=tCompras
    lTCompras.append(tTCompras)
    lCon.append(con)
    j+=1
for i in range(len(vArticulos) - 1):
    for j in range(len(vArticulos)):
        if(lCon[i] > lCon[j]):
            aux=lTCompras[i] 
            lTCompras[i]=lTCompras[j]
            lTCompras[j]=aux
            aux1=vIDart[i] 
            vIDart[i]=vIDart[j]
            vIDart[j]=aux1
            aux2=precioArt[i] 
            precioArt[i]=precioArt[j]
            precioArt[j]=aux2
            aux3=lCon[i] 
            lCon[i]=lCon[j]
            lCon[j]=aux3
            aux4=vArticulos[i] 
            vArticulos[i]=vArticulos[j]
            vArticulos[j]=aux4
            
for k in range(5):
    print("=========================================")
    vArticulos
    print("ID del artıculo -----> ",vIDart[k])
    print("Nombre del artıculo -----> ",vArticulos[k])
    print("Numero de compras ",lCon[k])
    print("Precio del artıculo ",precioArt[k])
    print("Valor total de las compras ",lTCompras[k])
    print("=========================================")
```
